```
import re
files = []
print(X)
for file in X:
  Y = re.split(r'[ ,\t]', file)
  files += [y for y in Y if y!=""]

print(len(files))
```

```
from collections import defaultdict
DataDict = defaultdict(lambda: {"C_text": str, "R_text": str , "Discourse_facet" : str})
c = 0
featureDict = {}
for rfile in files:
  try :
    open_file = open(rfile+"/annotation/"+rfile+".ann.txt", 'r')
  except:
    open_file = open(rfile+"/annotation/"+rfile+".annv3.txt", 'r')
  citances = open_file.readlines()
  citanceNumber = 0
  for citance in citances:
    citanceNumber = citanceNumber + 1
    if len(citance) > 1:
      data_file = citance.split(" | ")
      if(len(data_file) == 11):
        C_text = (data_file[6].split(":")[1])
        R_text = (data_file[8].split(":")[1])
        DataDict[data_file[1].split(":")[1]]["C_text"] = " ".join(C_text)
        DataDict[data_file[1].split(":")[1]]["R_text"] = " ".join(R_text)
        sentencepos = []
        sectionpos = []
        facetpos = []
        facetpos.append(-2)
        facetpos.append(-2)
        # Five facets 
        #  Aim, Hypothesis, Implication, Results, Method 
        mat = []
        C_textb = C_text.split("</S>")
        for entry in C_textb:
          if entry.find(">") != -1 :
            temp = entry.split(">")[0]  
            temp2 = temp.split("=")[1]
            temp3 = temp.split("=")[2]
            sentencepos.append((int)(temp2.split("\"")[1]))
            sectionpos.append((int)(temp3.split("\"")[1]))
        if len(sentencepos)<2  :
          sentencepos.append(-1)
          sectionpos.append(-1)
        for i in range(2) :
          vec = []
          vec.append(sentencepos[i])
          vec.append(sectionpos[i])
          vec.append(facetpos[i])
          mat.append(vec)
        featureDict[(rfile,citanceNumber)]=mat

        if(c==0):
          print(featureDict)
          print(sentencepos)
          print(sectionpos)
          print(data_file)
          print(C_textb)
          print(C_text)
          print(R_text)
          c=1;
  print(featureDict)
```

```
X = []
y = []
cnt = 0
for rfile in files:
  try :
    open_file = open(rfile+"/annotation/"+rfile+".ann.txt", 'r')
  except:
    open_file = open(rfile+"/annotation/"+rfile+".annv3.txt", 'r')
  citances = open_file.readlines()
  for citance in citances:
    if len(citance) > 1:
      data_file = citance.split(" | ")
      if(len(data_file) == 11):
        R_text = (data_file[8].split(":")[1])
        Facet = (data_file[9].split(":")[1])
        result = R_text.find('sid') 
        temp = (R_text[result:result+10])
        t1 = temp.split('"')[1]
        # print(t1)
        result = R_text.find('ssid') 
        temp = (R_text[result:result+11])
        t2 = 0
        try:
          t2 = temp.split('"')[1]
        except:
          t2 = 0
        # print(t2)
        t = [int(t1), int(t2)]
        # print(t)
        X.append(t)
        g = 0
        if (Facet.find('Method_Citation') != -1):
          cnt += 1
          g = 1
        y.append(g)
print(len(X),len(y),cnt)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import average_precision_score
from sklearn.model_selection import train_test_split

import numpy as np

X = np.array(X)
y = np.array(y)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42)

print(X_train.shape,y_test.shape)
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
# clf.predict(X_test)
clf.score(X_test,y_test)
```
